## Recommendation Model

```python
from surprise import Reader,Dataset
```

```python
amazon.columns
```

```python
melt_amazon=amazon.melt(id_vars=amazon.columns[0],value_vars=amazon.columns[1:],var_name="Movie_name",value_name="Rating")
melt_amazon
```

```python
reader=Reader(rating_scale=(-1,10))

data=Dataset.load_from_df(melt_amazon.fillna(0),reader=reader)
data
```

```python
from surprise.model_selection import train_test_split
```

```python
train,test=train_test_split(data,train_size=0.75,random_state=25)
```

```python
from surprise import SVD
```

```python
algo=SVD()
algo.fit(train)
```

```python
pred=algo.test(test)
pred
```

```python
from surprise import accuracy
```

```python
accuracy.rmse(pred)
```

```python
accuracy.mae(pred)
```

```python
from surprise.model_selection import cross_validate
```

```python
cv=cross_validate(algo,data,measures=["rmse","mae"],cv=3,verbose=True)
cv
```

```python
def repeat(algo_type,frame,min_,max_):
    reader=Reader(rating_scale=(-1,10))
   
    data=Dataset.load_from_df(frame,reader=reader)
    
    print(cross_validate(algo_type,data,measures=["RMSE","MAE"],cv=3,verbose=True))
    
    print("#"*10)
    user_id="A3R5OBKS7OM2IR"
    movie_id="Movie1"
    r_ui= 5.0
    print(algo_type.predict(user_id,movie_id,r_ui=r_ui,verbose = True))
    
    print("#"*10)
```

```python
repeat(SVD(),melt_amazon.fillna(melt_amazon["Rating"].mean()),-1,10)
```

```python
repeat(SVD(),melt_amazon.fillna(melt_amazon["Rating"].median()),-1,10)
```

```python
from surprise.model_selection import GridSearchCV
```

```python
parameter_grid={"n_epochs":[20,30],"lr_all":[0.005,0.01],"n_factors":[50,100]}
parameter_grid
```

```python
gs=GridSearchCV(SVD,parameter_grid,measures=["rmse","mae"],cv=3)

data1=Dataset.load_from_df(melt_amazon.fillna(melt_amazon["Rating"].mean()),reader=reader)

gs.fit(data1)
```
